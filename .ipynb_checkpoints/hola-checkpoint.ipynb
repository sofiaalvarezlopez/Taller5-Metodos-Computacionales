```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc 
```

```python
# Leer datos
data = np.loadtxt('data/ManchasSolares.dat')
```

```python
fecha = data[:,0] + data[:,1]/12. # escribir la fecha en unidades de a√±os
manchas = data[:,3]

print(fecha, manchas)
```

```python
fig = plt.figure(figsize=(10,5))
plt.subplot(1,2,1)

plt.xlabel(r'Years')
plt.ylabel(r'Frecuencia [1/mes]')

plt.plot(fecha, manchas, label=r'Manchas Solares')
```

```python
# cortar solo para los datos despues de 1900

ii = fecha > 1900
```

```python
# Quitar la media para que domine la frecuencia cero

mean=np.mean(manchas[ii])
manchas = manchas[ii] - mean
fecha = fecha[ii]
```

```python
fig = plt.figure(figsize=(10,5))
plt.subplot(1,2,1)

plt.xlabel(r'Years')
plt.ylabel(r'Frecuencia [1/mes]')


plt.plot(fecha, manchas+mean, label=r'Manchas Solares')
```

```python
# Calcular la trasformada de Fourier de los datos

Npoints = len(manchas)
FFT = np.fft.fft(manchas)
Frecuencias = np.fft.fftfreq(Npoints,1)
FFTA = np.abs(FFT)
```

```python
fig = plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(np.arange(Npoints), np.abs(FFT/Npoints))
plt.stem(np.arange(Npoints), np.abs(FFT/Npoints))
```

```python
# Encontrar frecuencia dominante
ii = np.argmax(FFTA)
periodo = (1./Frecuencias[ii])/12.0
```

```python
# Frecuencias mayor y menor a la dominante
ii_A = np.abs(Frecuencias) > (Frecuencias[ii])
ii_B = np.abs(Frecuencias) < (Frecuencias[ii])
```

```python
# Hacer cero todos los modos menos el de la frecuencia dominante
FFT[ii_A | ii_B] = 0.
```

```python
fig = plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(np.arange(Npoints), np.abs(FFT/Npoints))
plt.stem(np.arange(Npoints), np.abs(FFT/Npoints))
```

```python
# Calcular la transformada inverse
manchas_do = np.fft.ifft(FFT)
```

```python
fig = plt.figure(figsize=(7,5))
plt.subplot(1,1,1)
plt.title(r'Periodo {:.2f} en a$\~n$os'.format(periodo))
plt.plot(fecha, manchas+mean, label='Datos')
plt.plot(fecha, np.real(manchas_do)+mean, c='r', label='Frecuencia dominante')

plt.xlabel('Frecuencia [1/mes]')
plt.ylabel('Norma FFT')
plt.legend()
plt.savefig('PeriododeManchas.pdf')
```
